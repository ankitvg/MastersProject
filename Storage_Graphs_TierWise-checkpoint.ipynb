```python
import pandas as pd
import numpy as np
```

```python
dsFile = pd.read_csv('data_files/ds_dump_1.csv')
```

```python
import math
import matplotlib.pyplot as plt
```

```python
# all economy graphs for Jun 2016
def get_DiskUsage_Graph_Month(month):
    if month == "jun":
        month_str = '06\/[0-9][0-9]\/2016'
    elif month == "jul":
        month_str = '07\/[0-9][0-9]\/2016'
    elif month == "aug":
        month_str = '08\/[0-9][0-9]\/2016'
    elif month == "sep":
        month_str = '09\/[0-9][0-9]\/2016'
    elif month == "nov":
        month_str = '2016-11'
    elif month == "oct":
        month_str = '10\/[0-9]{1,2}\/2016'
    else:
        return "Month passed is not valid"
    ds_jun2016_all = dsFile[dsFile.ds_time.str.contains(month_str) > 0]
    ds_jun2016_all['ds_time'] = ds_jun2016_all['ds_time'].apply(lambda x: str(x).replace(str(x)[-5:],'00:00'))
    if month == "oct":
        ds_jun2016_all['Date'] = pd.to_datetime(ds_jun2016_all['ds_time'], format="%m/%d/%Y %H:%M:%S")
        ds_oct22016_all = dsFile[dsFile.ds_time.str.contains('2016-10') > 0]
        ds_oct22016_all['ds_time'] = ds_oct22016_all['ds_time'].apply(lambda x: str(x).replace(str(x)[-5:],'00:00'))
        ds_oct22016_all['Date'] = pd.to_datetime(ds_oct22016_all['ds_time'], yearfirst=True)
        ds_jun2016_all = ds_jun2016_all.append(ds_oct22016_all)
    elif month != "nov":
        ds_jun2016_all['Date'] = pd.to_datetime(ds_jun2016_all['ds_time'], format="%m/%d/%Y %H:%M:%S")
    elif month == "nov":
        ds_jun2016_all['Date'] = pd.to_datetime(ds_jun2016_all['ds_time'], yearfirst=True)
    else:
        return "Month passed is not valid"
    ds_jun2016_all['used_space'] = ds_jun2016_all['ds_cap_gb'] - ds_jun2016_all['ds_free_space']
    ds_jun2016_GB_graph_tier = ds_jun2016_all.groupby(['ds_storagetier', 'ds_loc', 'Date'])[["used_space"]].sum()
    storage_tiers = ["STANDARD", "ECONOMY", "HIGHPERFORMANCE"]
    for storageTier in storage_tiers:
        ds_jun2016_GB_graph_tier_eco =  ds_jun2016_GB_graph_tier.iloc[ds_jun2016_GB_graph_tier.index.get_level_values('ds_storagetier') == storageTier]
        unique_locs = ds_jun2016_GB_graph_tier_eco.index.get_level_values('ds_loc').unique()
        ncols=5
        nrows=math.ceil(unique_locs.size/5)
        m=0
        n=0
        fig, ax = plt.subplots(nrows,ncols)
        fig.set_figheight(10)
        fig.set_figwidth(10)
        fig.suptitle(storageTier  + " storage tier for " + month + " 2016")
        fig.subplots_adjust(hspace=.5)
        for location in unique_locs:
            temp = ds_jun2016_GB_graph_tier_eco.iloc[ds_jun2016_GB_graph_tier_eco.index.get_level_values('ds_loc') == location]
            temp=temp.reset_index(level=['Date'])
            temp2=pd.DataFrame()
            temp2['Date']=temp['Date']
            temp2['used_space']=temp['used_space']
            temp2 = temp2.set_index('Date')
            temp2.to_csv("temp2_"+location ,sep='\t')
            ts=(temp2['used_space'])
            ax[m,n].plot(ts, marker='.')
            ff=location+" "+str(temp2.size)
            ax[m,n].set_title(ff)
            ax[m,n].axes.get_xaxis().set_visible(False)
            ax[m,n].axes.get_yaxis().set_visible(False)
            if n==4:
                n=0
                m+=1
            else:
                n+=1
        plt.show()
```

```python
get_DiskUsage_Graph_Month("nov")
```

```python
for month in ["jul", "aug", "sep"]:
    get_DiskUsage_Graph_Month(month)
```
